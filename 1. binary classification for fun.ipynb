### 4. Data Preprocessing

#### Train-test data split

- Before we do imputation on missing values, I'll first devide the dataset into train and test set to avoid data leakage. Here, I will do stratified random sampling to make sure target variables have same ratio in both datasets.

```python
from sklearn.model_selection import train_test_split
X = df.drop('y',axis=1)
y = df['y']
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.15, stratify=y, random_state=42)
```

```python
plt.figure(figsize=(6,3))
plt.subplot(1,2,1)
sns.countplot(y_train)
plt.xlabel('train')
plt.ylabel('counts')

plt.subplot(1,2,2)
sns.countplot(y_test)
plt.xlabel('test')
plt.ylabel('counts')
plt.tight_layout();
```

#### Imputation of missing values

```python
from sklearn.impute import KNNImputer, SimpleImputer
imput_num = KNNImputer()
imput_cat = SimpleImputer(strategy='most_frequent')

# impute numeric features using 'KNN imputation' and categorical features using 'most frequent' imputation
x_train_num = x_train.drop(['x34','x35','x68','x93'], axis=1).copy()
x_train_cat = x_train[['x34','x35','x68','x93']].copy()

x_train_num = pd.DataFrame(imput_num.fit_transform(x_train_num),
                                   columns=x_train_num.columns, index=x_train_num.index)
x_train_cat = pd.DataFrame(imput_cat.fit_transform(x_train_cat),
                                   columns=x_train_cat.columns, index= x_train_cat.index)

# make sure there's no null value
print(x_train_num.isnull().sum().sum())
print(x_train_cat.isnull().sum().sum())

# combine num and cat columns
x_train = pd.merge(x_train_num, x_train_cat, left_index=True, right_index=True).sort_index()
print(x_train.shape)
```

#### Creating new features

```python
# deal with cyclic features
x_train['x68_2'] = x_train.x68.replace(
                          ['Jul','Jun','Aug','Sep','May','Apr','Mar','Dec','Oct','Nov','Feb','Jan'],
                          [7,6,8,9,5,4,3,12,10,11,2,1])

# sin and cos transformation of cyclic features
x_train['x68_sin'] = np.sin((x_train['x68_2']-1)*(2.*np.pi/12))
x_train['x68_cos'] = np.cos((x_train['x68_2']-1)*(2.*np.pi/12))
x_train.drop('x68_2', axis=1, inplace=True)

x_train['x35_2'] = x_train.x35.replace(['wed', 'thur', 'fri', 'tue', 'mon'], [3,4,5,2,1])
x_train['x35_sin'] = np.sin((x_train['x35_2']-1)*(2.*np.pi/5))
x_train['x35_cos'] = np.cos((x_train['x35_2']-1)*(2.*np.pi/5))
x_train.drop('x35_2', axis=1, inplace=True)
```

#### Categorical encoding and feature scaling

- Next, I will encode categorical variables as dummy variables and scale the data using minmax scaler.

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
x_train_num = x_train.drop(['x34','x35','x68','x93'], axis=1)
x_train_cat = x_train[['x34','x35','x68','x93']]

# onehot encoding categorical variables
onehot = OneHotEncoder()
onehot.fit(x_train_cat)
x_train_dum = pd.DataFrame(onehot.transform(x_train_cat).toarray(), 
                           columns=onehot.get_feature_names(), index=x_train_cat.index)
train = pd.merge(x_train_num, x_train_dum, left_index=True, right_index=True)

# minmax scaling
scale = MinMaxScaler()
scale.fit(train)
train = pd.DataFrame(scale.transform(train), columns=train.columns, index=train.index)
```

```python
train.shape
```

```python
# Function to perform the same preprocessing to the test dataset.
def preprocessing(df):

    # imputate missing value
    df_num = df.drop(['x34','x35','x68','x93'], axis=1)
    df_cat = df[['x34','x35','x68','x93']]
    df_num = pd.DataFrame(imput_num.transform(df_num), columns=df_num.columns, index=df_num.index)
    df_cat = pd.DataFrame(imput_cat.transform(df_cat), columns=df_cat.columns, index=df_cat.index)
    df = pd.merge(df_num, df_cat, left_index=True, right_index=True).sort_index()
    # add cyclic feature
    df['x35_2'] = df.x35.replace(['wed', 'thur', 'fri', 'tue', 'mon'], [3,4,5,2,1])
    df['x68_2'] = df.x68.replace(['Jul','Jun','Aug','Sep','May','Apr','Mar','Dec','Oct','Nov','Feb','Jan'],
                                 [7,6,8,9,5,4,3,12,10,11,2,1])
    df['x35_sin'] = np.sin((df['x35_2']-1)*(2.*np.pi/5))
    df['x35_cos'] = np.cos((df['x35_2']-1)*(2.*np.pi/5))
    df['x68_sin'] = np.sin((df['x68_2']-1)*(2.*np.pi/12))
    df['x68_cos'] = np.cos((df['x68_2']-1)*(2.*np.pi/12))  
    df.drop('x35_2', axis=1, inplace=True)
    df.drop('x68_2', axis=1, inplace=True)    
    # transform categorical variables to dummy variables
    cat = df[['x34','x35','x68','x93']].copy()
    num = df.drop(['x34','x35','x68','x93'], axis=1)
    dum = pd.DataFrame(onehot.transform(cat).toarray(), columns=onehot.get_feature_names(), index=cat.index)
    df = pd.merge(num, dum, left_index=True, right_index=True)  
    # scaler
    df = pd.DataFrame(scale.transform(df), columns=df.columns, index=df.index).sort_index()
    return df
```

```python
test = preprocessing(x_test)
train, test, y_train, y_test = train.sort_index(), test.sort_index(), y_train.sort_index(), y_test.sort_index()
```

```python
print(train.shape)
print(test.shape)
```

#### Use undersampling to deal with imbalance

```python
from imblearn.under_sampling import RandomUnderSampler
rus = RandomUnderSampler(random_state=0)
X_resampled, y_resampled = rus.fit_resample(train, y_train)
```

## III. Model building

```python
from sklearn.metrics import roc_auc_score, confusion_matrix
from sklearn.model_selection import GridSearchCV
```

```python
def evaluate(classifier, test=test, y_test=y_test):
    
    y_pred = classifier.predict(test)
    print(confusion_matrix(y_test, y_pred))
    print('auc score is:', round(roc_auc_score(y_test, y_pred),4))    
```

### Model 1: logistic regression

- First, I will use entire dataset to fit the model and use cost-sensitive learning (by giving different class weight in the model) to deal with imbalance.  Then I'll compare the result with the model using undersampled data to pick final model to use.

#### 1. Logistic regression: cost-sensitive learning with hyperparameter tuning using grid search

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(class_weight={0:0.2, 1:0.8})
lr_param = {'penalty': ['l1','l2','elastic_net'], 
            'solver': ['liblinear'], 
            'C': [0.001,0.01,0.1,1,10,100]}
lr_grid = GridSearchCV(lr, lr_param, scoring='roc_auc', cv=5, verbose=2, n_jobs = -1)       

lr_grid.fit(train, y_train)
```

```python
# Validation set auc score of the best estimator is:
evaluate(lr_grid, test, y_test)
```

```python
# Training set auc score of the best estimator is:
evaluate(lr_grid, train, y_train)
```

- Very similar train and validation auc scores mean that there's no overfitting going on, which is good.

#### 2. Logistic regression: undersampled data with hyperparameter tuning using grid search.

```python
lr_us = LogisticRegression()
lr_param = {'penalty': ['l1','l2','elastic_net'], 
            'solver': ['liblinear'], 
            'C': [0.001,0.01,0.1,1,10,100]}
lr_grid_us = GridSearchCV(lr_us, lr_param, scoring='roc_auc', cv=5, verbose=2, n_jobs = -1)       
lr_grid_us.fit(X_resampled, y_resampled)
```

```python
# Validation set auc score is:
evaluate(lr_grid_us, test, y_test)
```

```python
# Train set auc score is:
evaluate(lr_grid_us, train, y_train)
```

- Because first logistic regression model (cost-sensitive learning) shows slightly higher validation auc score, I'll use it as final model.
